```python
import nltk
nltk.download('punkt')
```

```python
import os
from nltk.tokenize import word_tokenize

class MinhasSentencas(object):
    def __init__(self, pasta):
        self.pasta = pasta
 
    def __iter__(self):
        for arq in os.listdir(self.pasta):
            texto = ''
            for linha in open(os.path.join(self.pasta, arq), 'r'):
                texto += linha.lower()
            tokens = word_tokenize(texto)
            tokens_ok = []
            for token in tokens:
                if len(token) <= 100:
                    tokens_ok.append(token)
            yield tokens_ok
```

```python
import gensim, logging
import time
import multiprocessing
from gensim.models import Word2Vec

logging.basicConfig(format='%(asctime)s : %(levelname)s : %(message)s', level=logging.INFO)
cores = multiprocessing.cpu_count()
PASTA_ENTRADA = '../externos/acordaos-todos/'

print('Usando', cores-1, 'workers.')

sentences = MinhasSentencas(PASTA_ENTRADA)
model = Word2Vec(sentences,
                 max_vocab_size=50000,
                 min_count=20,
                 window=2,
                 size=100,
                 sample=6e-5, 
                 alpha=0.03, 
                 min_alpha=0.0007, 
                 negative=20,
                 workers=cores-1)
```
