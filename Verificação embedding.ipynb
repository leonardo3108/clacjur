```python
import pandas as pd

df = pd.read_csv('jurisprudencia_selecionada_enunciados.txt', sep = '|')
df.head()
```

```python
from keras.preprocessing.text import Tokenizer
import numpy as np

limite_texto = 200
dim_vetor = 100

tokenizer = Tokenizer()
tokenizer.fit_on_texts(df['TEXTO_ENUNCIADO'])

word_index = tokenizer.word_index
vocabulario = len(word_index) + 1
print('Found %s unique tokens.' % len(word_index))

sequences = tokenizer.texts_to_sequences(df['TEXTO_ENUNCIADO'])
```

```python
tokenizer.word_index
```

```python
freq = dict()
for sentence in sequences:
    for word in sentence:
        if word in freq:
            freq[word] += 1
        else:
            freq[word] = 1
```

```python
tops = sorted(freq.values(), reverse=True)[:10]
tops
```

```python
for k in freq.keys():
    if freq[k] in tops:
        print(k, freq[k])
```

```python
for word in tokenizer.word_index.items():
    print(word[0], freq[word[1]])
```

```python
from gensim.models import KeyedVectors

model = KeyedVectors.load_word2vec_format('model.txt')
```

```python
# create a weight matrix for words in training docs

embedding_matrix = np.zeros((vocabulario, 100))

ok = 0
nok = []
for word, i in tokenizer.word_index.items():
    if word in model:
        embedding_matrix[i] = model[word]
        ok += 1
    else:
        nok.append((word,i))
print('Vocabulario:', i)
print('Encontrados no modelo:', ok, '=', ok * 100. / i)
```

```python
qtde = 0
for word in nok:
    f = freq[word[1]]
    qtde += f
    print (word[0], f)
print('Total', qtde, '; distintas', len(nok))
```
